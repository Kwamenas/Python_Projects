```python
def remove_binary_col(df):
    binary_only= df.nunique()==2
    binary_cols=binary_only[binary_only].index.tolist()
    return df.drop(columns=binary_cols)
    
    
    
def rows_with_max_col (df):
    max_cols=df.idxmax()
    max_cols=max_cols.unique()
    return df.loc[max_cols]
    
    
```
