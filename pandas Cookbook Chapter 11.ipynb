```python
import sqlalchemy as sa
```

```python
engine=sa.create_engine('sqlite:///C:\\Users\\287910031\\9781839213106_Code\\data\\chinook.db',echo=True)
```

```python
tracks=pd.read_sql_table('tracks',engine)
```

```python
tracks
```

```python
genre=pd.read_sql_table('genres',engine)
```

```python
genre
```

```python
tracks.columns
```

```python
(genre.merge(tracks,on='GenreId',how='left',suffixes=('_gen','_track'))
 [['GenreId','Milliseconds','Name_gen']]).drop('GenreId',axis='columns')
```

```python
((genre.merge(tracks,on='GenreId',how='left',suffixes=('_gen','_track'))
 [['GenreId','Milliseconds','Name_gen']]).drop('GenreId',axis='columns')
 .groupby('Name_gen')['Milliseconds'].agg('mean')
 .pipe(lambda s:pd.to_timedelta(s,unit='ms'))
 .rename('Length')
 .dt.floor('s')
 .sort_values()
)
```

```python
customers=pd.read_sql_table('customers',engine)
invoice=pd.read_sql_table('invoices',engine)
invoice_items=pd.read_sql_table('invoice_items',engine)
```

```python
customers.columns
```

```python
invoice.columns
```

```python
invoice_items.columns
```

```python
(customers[['CustomerId','FirstName','LastName']].merge(invoice[['InvoiceId','CustomerId']],on='CustomerId')
 .merge(invoice_items[['InvoiceId', 'UnitPrice', 'Quantity']],on='InvoiceId'))
```

```python
(customers[['CustomerId','FirstName','LastName']].merge(invoice[['InvoiceId','CustomerId']],on='CustomerId')
 .merge(invoice_items[['InvoiceId', 'UnitPrice', 'Quantity']],on='InvoiceId')
 .assign(Total=lambda df: df['Quantity'] * df['UnitPrice'])
 .groupby(['CustomerId','FirstName','LastName'])['Total'].agg('sum')
 .sort_values(ascending=False))
```
